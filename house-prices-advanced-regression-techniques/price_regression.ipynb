```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score, GridSearchCV


```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
x_train = df_train.drop(columns=['SalePrice'])
y_train = df_train[['SalePrice']]
```

```python
y_train
```

```python
numeric_features = x_train.select_dtypes(include=[np.number]).columns
categorical_features = x_train.select_dtypes(include=[object]).columns
```

```python
numeric_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')), ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessing = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
      ('cat', categorical_transformer, categorical_features)
    ])
```

```python
tree_model = Pipeline(steps=[('preprocessor', preprocessing),
                             ('feature_selection', SelectKBest(score_func=f_regression, k=13)),
                             ('tree', DecisionTreeRegressor())])
```

```python
cv_scores = cross_val_score(tree_model, x_train, y_train, cv=7, scoring='neg_mean_squared_error')
grid_search = GridSearchCV(tree_model, param_grid={'feature_selection__k': range(1, 15)}, cv=7, scoring='neg_mean_squared_error')

grid_search.fit(x_train, y_train)
cv_scores = cross_val_score(grid_search.best_estimator_, x_train, y_train, cv=7, scoring='neg_mean_squared_error')

print("Best model: {}".format(grid_search.best_params_))
```
